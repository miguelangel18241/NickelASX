```python
# List of ASX ticker codes for nickel companies
asx_ticker_codes = ["WSA.AX", "IGO.AX", "PAN.AX", "MCR.AX", "POS.AX"]
indices_to_remove = [0, 3]

asx_ticker_codes = [ticker for i, ticker in enumerate(asx_ticker_codes) if i not in indices_to_remove]

# Define the start and end dates for the historical data
start_date = "2010-02-27"
end_date = "2024-02-27"

## Get from ASX

historical_data = {}

for ticker_code in asx_ticker_codes:
  ##Download Data
  data = yf.download(ticker_code, start = start_date, end = end_date)
  ##Store it
  historical_data[ticker_code] = data

#Dic to Store the DataFrame objects:

dataframes = {}

# Convert the historical data for each ticker into a DataFrame

for ticker_code, data in historical_data.items():
  df = pd.DataFrame(data)
  dataframes[ticker_code] = df
```

```python
# List of ASX ticker codes for nickel companies
asx_ticker_codes = ["WSA.AX", "IGO.AX", "PAN.AX", "MCR.AX", "POS.AX"]
indices_to_remove = [0, 3]

asx_ticker_codes = [ticker for i, ticker in enumerate(asx_ticker_codes) if i not in indices_to_remove]

# Define the start and end dates for the historical data
start_date = "2010-02-27"
end_date = "2022-11-29"

## Get from ASX

historical_data = {}

for ticker_code in asx_ticker_codes:
  ##Download Data
  data = yf.download(ticker_code, start = start_date, end = end_date)
  ##Store it
  historical_data[ticker_code] = data

#Dic to Store the DataFrame objects:

dataframes = {}

# Convert the historical data for each ticker into a DataFrame

for ticker_code, data in historical_data.items():
  df = pd.DataFrame(data)
  dataframes[ticker_code] = df

# Print the first few rows of the DataFrame corresponding to the first ticker code
first_ticker_code = asx_ticker_codes[0]
first_dataframe = dataframes[first_ticker_code].reset_index()
print("First few rows of historical data for", first_ticker_code)
print(first_dataframe.tail())
```

```python
average_volume = first_dataframe['Volume'].mean()
max_volume = first_dataframe['Volume'].max()
increase_percentage = ((max_volume-average_volume)/average_volume)*100

```

```python
print(f'''Average is: {round(average_volume,2)} and highest volume is: {max_volume}, today's volumen was higher by
 {round(increase_percentage,2)}%''')
```
